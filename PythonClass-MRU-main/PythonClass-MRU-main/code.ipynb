```python
# variable length arguments :- collect all the variables that are not in use(*)
def sum(a,b,*c):
     return f"{a+b} and unused values are{c} "
print(sum(5,10,5))
print(sum(5,200,500,120))

# keyword length arguments:- collect all the keyword arguments that are not in use(**)
def greetings(name,**temp):
     print("Welcome ",name)
     print(temp)
greetings("python",id="101",age=40)
# lambda function :- a function without name and it used for one time syntax: lambda arguments:expression or operation
n = int(input("enter n value"))
r = lambda x:x*x
print(r(n))
# ternary operator (result1 if condition else result2)
a = 10
b = 5
r = f"a value {a} is largest num" if a > b else  f"b value {b} is largest num"
print(r)
# map is used to iterate a sequence and apply some condition on each element
l = [5,10,15,20]
# without map
m = [];
for i in l:
     m.append(i*5)
print(m)
# map 
r = list(map(lambda x:x*5,l))
print(r)
# filter is used to apply some condition on sequence filter(func,sequence)
r = list(filter(lambda x: x% 2 != 0,l))
print(r)
# reduce function
# import functools
from functools import reduce

l = [5,10,15,20]
r = reduce(lambda x,y:x+y,l)
print(r)

```

```python
# api calls in python
import requests
API_URL = 'https://fakestoreapi.com/products/'
try:
     response = requests.get(API_URL)
     print(response)
     if response.status_code == 200:
          data = response.json()
          print(data)
     else:
          print("404 not found")
except Exception as e:
     print(e)
```
